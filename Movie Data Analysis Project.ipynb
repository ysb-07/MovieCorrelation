```python
# Import Libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

# Read in the data

df=pd.read_csv(r'E:\Project\Project 3 - Movie Dataset\movies.csv')
```

```python
# Let's look at the data

df.head()
```

```python
#Checking for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,round(pct_missing*100)))
```

```python
df = df.dropna()
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,round(pct_missing*100)))
```

```python
df.head()
```

```python
df.dtypes
```

```python
#Changing Data Types

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
df['votes'] = df['votes'].astype('int64')
df['runtime'] = df['runtime'].astype('int64')
```

```python
df.head()
```

```python
#Create Correct year column
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
df.head()
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
df.head()
```

```python
# Check for Duplicates

df.drop_duplicates()

df.head()
```

```python
# Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget for Film')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
df.head()
```

```python
# Plot the budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df,scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# Correlation

df.corr(method = 'pearson')
```

```python
correlation_matrix = df.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerzied = df
for col_name in df_numerzied.columns:
    if(df_numerzied[col_name].dtype) == 'object':
        df_numerzied[col_name] = df_numerzied[col_name].astype('category')
        df_numerzied[col_name] = df_numerzied[col_name].cat.codes
df_numerzied.head()
```

```python
correlation_matrix = df.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
correlation_mat=df_numerzied.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[sorted_pairs>0.5]
high_corr
```
